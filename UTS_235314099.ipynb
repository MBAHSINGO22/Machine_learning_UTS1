```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

from yellowbrick.cluster import KElbowVisualizer
import scipy.cluster.hierarchy as sch
```

```python
# Load data
df = pd.read_csv("WineQT.csv")
```

```python
# Preprocessing 
# Drop duplicates
df = df.drop_duplicates()
```

```python
# Cek dan hilangkan kolom non-numerik jika ada
df = df.select_dtypes(include=[np.number])
```

```python
# Feature Scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
df.iloc[:,0:]
df
```

```python
# PCA (for visualization)
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
```

```python
#KMeans Clustering 
print("\n--- KMEANS ---")
model_kmeans = KMeans()
visualizer = KElbowVisualizer(model_kmeans, k=(2,10))
visualizer.fit(scaled_data)
visualizer.show()
```

```python
# Hierarchical Clustering
print("\n--- HIERARCHICAL ---")
# Dendrogram untuk menentukan jumlah cluster
plt.figure(figsize=(10, 6))
dendrogram = sch.dendrogram(sch.linkage(scaled_data, method='ward'))
plt.title("Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

```python
#DBSCAN Clustering
print("\n--- DBSCAN ---")
dbscan = DBSCAN(eps=1.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(scaled_data)
# DBSCAN bisa menghasilkan -1 untuk outlier
n_clusters_dbscan = len(set(labels_dbscan)) - (1 if -1 in labels_dbscan else 0)
print("Jumlah cluster DBSCAN:", n_clusters_dbscan)
```

```python
# Hitung silhouette score
sil_kmeans = silhouette_score(scaled_data, labels_kmeans)
sil_hier = silhouette_score(scaled_data, labels_hier)

# Untuk DBSCAN, hanya hitung jika ada lebih dari 1 cluster
if len(set(labels_dbscan)) > 1:
    sil_dbscan = silhouette_score(scaled_data, labels_dbscan)
else:
    sil_dbscan = None

# Visualisasi gabungan dengan Silhouette Score
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# KMeans
axes[0].scatter(pca_data[:, 0], pca_data[:, 1], c=labels_kmeans, cmap='rainbow')
axes[0].set_title(f'KMeans Clustering\nSilhouette Score: {sil_kmeans:.2f}')
axes[0].set_xlabel('PCA 1')
axes[0].set_ylabel('PCA 2')

# Hierarchical
axes[1].scatter(pca_data[:, 0], pca_data[:, 1], c=labels_hier, cmap='viridis')
axes[1].set_title(f'Hierarchical Clustering\nSilhouette Score: {sil_hier:.2f}')
axes[1].set_xlabel('PCA 1')
axes[1].set_ylabel('PCA 2')

# DBSCAN
axes[2].scatter(pca_data[:, 0], pca_data[:, 1], c=labels_dbscan, cmap='plasma')
dbscan_title = f'DBSCAN Clustering\nSilhouette Score: {sil_dbscan:.2f}' if sil_dbscan else "DBSCAN Clustering\nSilhouette Score: N/A"
axes[2].set_title(dbscan_title)
axes[2].set_xlabel('PCA 1')
axes[2].set_ylabel('PCA 2')

plt.tight_layout()
plt.show()
```
